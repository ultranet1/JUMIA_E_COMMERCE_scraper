```
i=1
pl=[]
while i <= 50:
  p='page='.strip() + str(i) 
  p1='https://www.jumia.com.ng/laptops/?'.strip()+str(p)
  print(p1)

  i=i+1
  pl.append(p1)
```

```
links=pl[0:50]

name=[]
```

```

for i in links:
  wd.get(str(i))
  p=wd.find_elements_by_class_name('name')
  
  try:
    for t in p:
      print(t.text)
      name.append(t.text)
  
  except:
    wd.implicitly_wait(5)
    pass
```

```
print(len(name))
```

```
price=[]
for i in links:
  wd.get(str(i))
  wd.implicitly_wait(5)
  p=wd.find_elements_by_class_name('prc')
  wd.implicitly_wait(5)
  try:
    for t in p:
      print(t.text)
      price.append(t.text)
  except:
    pass
```
